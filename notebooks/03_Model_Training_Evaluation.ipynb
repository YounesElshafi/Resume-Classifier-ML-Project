## ü§ñ Model Training & Evaluation for Resume Classification

### 1. Import Required Libraries

```python
import pandas as pd
import numpy as np
import joblib
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import MultinomialNB
from sklearn.multiclass import OneVsRestClassifier
from sklearn.svm import SVC
import seaborn as sns
import matplotlib.pyplot as plt
```

### 2. Load Preprocessed Data

```python
X_train = joblib.load('../models/X_train.pkl')
X_test = joblib.load('../models/X_test.pkl')
y_train = joblib.load('../models/y_train.pkl')
y_test = joblib.load('../models/y_test.pkl')
```

### 3. Model Evaluation Function

```python
def evaluate_model(model, name="Model"):
    """
    Train the model and evaluate it using accuracy, classification report, and confusion matrix.
    """
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    
    acc = accuracy_score(y_test, y_pred)
    print(f"\nüìä {name} Accuracy: {acc:.4f}")
    
    print("\nüìù Classification Report:")
    print(classification_report(y_test, y_pred))
    
    cm = confusion_matrix(y_test, y_pred)
    plt.figure(figsize=(8, 6))
    sns.heatmap(cm, annot=True, fmt='d', cmap="Blues", xticklabels=np.unique(y_test), yticklabels=np.unique(y_test))
    plt.title(f"{name} - Confusion Matrix")
    plt.xlabel('Predicted')
    plt.ylabel('Actual')
    plt.tight_layout()
    plt.show()
```

### 4. Train & Evaluate Multiple Models

```python
evaluate_model(LogisticRegression(max_iter=1000), "Logistic Regression")
```

```python
evaluate_model(RandomForestClassifier(n_estimators=100, random_state=42), "Random Forest")
```

```python
evaluate_model(MultinomialNB(), "Naive Bayes")
```

```python
evaluate_model(KNeighborsClassifier(), "K-Nearest Neighbors")
```

```python
evaluate_model(OneVsRestClassifier(SVC()), "One-vs-Rest (SVC)")
```

```python
evaluate_model(OneVsRestClassifier(KNeighborsClassifier()), "One-vs-Rest (KNN)")
```
